```python
# importing required libraries
import re
import nltk
from nltk.corpus import movie_reviews
import string
import pandas as pd
from nltk.corpus import stopwords

nltk.download('movie_reviews')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
from numpy import zeros
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
docs = [movie_reviews.raw(i) for i in movie_reviews.fileids()]
docs[0]
```

1. Create a bag of words representation of the movie review data from the Text Mining Assignment 
- Use sklearn's CountVectorizer for the bag of words representation
- Create versions of the data using unigrams and also bigrams (ngram_range parameter in CountVectorizer allows you to specify what ngrams to use)

```python
vectorizer = CountVectorizer(stop_words='english')
unigrams = vectorizer.fit_transform(docs)
print("First 50 unigrams are:")
print(vectorizer.get_feature_names_out()[:50])
print("\nUnigram Bag of Words Representation:")
print(unigrams.toarray())
print("\nUnigram Shape:", unigrams.shape)
```

```python
vectorizer2 = CountVectorizer(stop_words='english',ngram_range=(2, 2))
bigrams = vectorizer2.fit_transform(docs)
print("First 50 bigrams are:")
print(vectorizer2.get_feature_names_out()[:50])
print("\nBigram Bag of Words Representation:")
print(bigrams.toarray())
print("\nBigram Shape:", bigrams.shape)
```

2. Create a tf-idf representation of the same movie review data from the (e.g., using sklearn's TfidfVectorizer)
- Create versions of the data using unigrams and also bigrams (ngram_range parameter in TfidfVectorizer allows you to specify what ngrams to use)

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_unigrams = tfidf_vectorizer.fit_transform(docs)
print("First 50 unigrams are:")
print(tfidf_vectorizer.get_feature_names_out()[:50])
print("\nUnigram Bag of Words Representation:")
print(tfidf_unigrams.toarray())
print("\nUnigram Shape:", tfidf_unigrams.shape)
```

```python
tfidf_vectorizer2 = TfidfVectorizer(stop_words='english',ngram_range=(2, 2))
tfidf_bigrams = tfidf_vectorizer2.fit_transform(docs)
print("First 50 bigrams are:")
print(tfidf_vectorizer2.get_feature_names_out()[:50])
print("\nBigram Bag of Words Representation:")
print(tfidf_bigrams.toarray())
print("\nBigram Shape:", tfidf_bigrams.shape)
```

3. Use Bayesian classification to classify reviews into positive and negative using both the unigram and bigram versions of the bag of words and tf-idf representations from 1 and 2 above (e.g., using sklearn's MultinomialNB or nltk's NaiveBayesClassifier)

```python
movie_reviews.categories()
```

We can see that, our movie review dataset has 2 target category, Negative and Positive.

```python
neg = movie_reviews.fileids('neg')
pos = movie_reviews.fileids('pos')
neg_reviews = [movie_reviews.raw(fileids=[i]) for i in neg]
pos_reviews = [movie_reviews.raw(fileids=[i]) for i in pos]
labels = [0] * len(neg_reviews) + [1] * len(pos_reviews)
train_data, test_data, train_labels, test_labels = train_test_split(docs, labels, test_size=0.2)
```

 Bayesian classification using CountVectorizer




```python
train_unigrams = vectorizer.fit_transform(train_data)
test_unigrams = vectorizer.transform(test_data)
unigramNB = MultinomialNB()
unigramNB.fit(train_unigrams, train_labels)
unigram_pred = unigramNB.predict(test_unigrams)
unigram_acc = accuracy_score(test_labels, unigram_pred)
print("Unigram bag of words accuracy using CountVectorizer :", unigram_acc)
train_bigrams = vectorizer2.fit_transform(train_data)
test_bigrams = vectorizer2.transform(test_data)
bigramNB = MultinomialNB()
bigramNB.fit(train_bigrams, train_labels)
bigram_pred = bigramNB.predict(test_bigrams)
bigram_acc = accuracy_score(test_labels, bigram_pred)
print("Bigram bag of words accuracy usnig CountVectorizer:", bigram_acc)
```

 Bayesian classification using TF-IDF


```python
train_tfidf_unigrams = tfidf_vectorizer.fit_transform(train_data)
test_tfidf_unigrams = tfidf_vectorizer.transform(test_data)
tfidf_unigramNB = MultinomialNB()
tfidf_unigramNB.fit(train_tfidf_unigrams, train_labels)
tfidf_unigram_pred = tfidf_unigramNB.predict(test_tfidf_unigrams)
tfidf_unigram_acc = accuracy_score(test_labels, tfidf_unigram_pred)
print("Unigram bag of words accuracy using TFIDF :", tfidf_unigram_acc)
tfidf_train_bigrams = tfidf_vectorizer2.fit_transform(train_data)
tfidf_test_bigrams = tfidf_vectorizer2.transform(test_data)
tfidf_bigramNB = MultinomialNB()
tfidf_bigramNB.fit(tfidf_train_bigrams, train_labels)
tfidf_bigram_pred = tfidf_bigramNB.predict(tfidf_test_bigrams)
tfidf_bigram_acc = accuracy_score(test_labels, tfidf_bigram_pred)
print("Bigram bag of words accuracy usnig TFIDF:", tfidf_bigram_acc)
```

4. Use nltk.collocations functions to calculate the top 10 bigram collocations based on PPMI

```python
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
tokenized_docs=[nltk.word_tokenize(i) for i in docs]
finder = BigramCollocationFinder.from_documents(tokenized_docs)
top_10=finder.nbest(bigram_measures.pmi, 10)
```

```python
from __future__ import print_function
import pyLDAvis

import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
# Load 20 newsgroups dataset
# First, the 20 newsgroups dataset available in sklearn is loaded. As always, the headers, footers and quotes are removed.
```

We are given catergories 'sci.med', 'sci.space', 'talk.politics.guns'.

```python
cat=[ 'sci.med', 'sci.space', 'talk.politics.guns']
newsgroups = fetch_20newsgroups(subset='all', categories= cat, remove=('headers', 'footers', 'quotes'))
docs_raw = newsgroups.data
print(len(docs_raw))
```

```python
from pprint import pprint
pprint(list(newsgroups.target_names))
```

```python
# Convert to document-term matrix
# Next, the raw documents are converted into document-term matrix, possibly as raw counts or in TF-IDF form.

tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 10)
dtm_tf = tf_vectorizer.fit_transform(docs_raw)
print(dtm_tf.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
print(dtm_tfidf.shape)
```

```python
# Fit Latent Dirichlet Allocation models - this step takes a while!

# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=20, random_state=0)
lda_tf.fit(dtm_tf)

# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=20, random_state=0)
lda_tfidf.fit(dtm_tfidf)
LatentDirichletAllocation(batch_size=128, doc_topic_prior=None,
             evaluate_every=-1, learning_decay=0.7,
             learning_method='online', learning_offset=10.0,
             max_doc_update_iter=100, max_iter=10, mean_change_tol=0.001,
             n_jobs=1, n_components=20, perp_tol=0.1, random_state=0,
             topic_word_prior=None, total_samples=1000000.0, verbose=0)
```

```python
# Visualizing the models with pyLDAvis, using the term frequency (tf) vectors
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer)
```

```python
# Visualizing the models with pyLDAvis, using the tf-idf vectors

pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tfidf_vectorizer)
```
